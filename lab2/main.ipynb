```python
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y_true = 2 * x
errors = np.random.normal(0, 2, size=x.shape)
y_measured = y_true + errors

plt.plot(x, y_true, label='Функція y=2x')
plt.scatter(x, y_measured, color='red', label='Виміряні дані')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Лінійна функція з помилками вимірювань')
plt.show()

mae = np.mean(np.abs(y_true - y_measured))
mse = np.mean((y_true - y_measured) ** 2)
print(f'MAE: {mae}')
print(f'MSE: {mse}')

results_df = pd.DataFrame({
    'X': x,
    'True Y': y_true,
    'Measured Y': y_measured,
    'Error': y_measured - y_true
})
results_df.to_csv('results.csv', index=False)
```
