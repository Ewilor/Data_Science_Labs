```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

np.random.seed(42)

X_numpy = np.linspace(0, 10, 100)

true_k, true_b = 3, 7

noise = np.random.normal(0, 2, 100)
y_numpy = true_k * X_numpy + true_b + noise

df_numpy = pd.DataFrame({'X': X_numpy, 'y': y_numpy})
print(df_numpy.head())

X_reshaped = df_numpy[['X']] 
y = df_numpy['y']

X_train, X_test, y_train, y_test = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"MAE: {mae:.2f} $")
print(f"RMSE: {rmse:.2f} $")
print(f"R2 Score: {r2:.2f}")

print(f"Справжня формула: y = {true_k} * x + {true_b}")
print(f"Модель знайшла:   y = {model.coef_[0]:.2f} * x + {model.intercept_:.2f}")

plt.scatter(X_numpy, y_numpy, color='green', alpha=0.5, label='Згенеровані дані')
plt.plot(X_numpy, model.predict(X_reshaped), color='red', linewidth=2, label='Знайдена лінія')
plt.title('Завдання 3+4: Numpy Generation')
plt.legend()
plt.show()
```
